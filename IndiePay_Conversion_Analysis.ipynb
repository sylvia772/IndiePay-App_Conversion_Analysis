FUNNEL ANALYSIS

```python
### creating the funnel


# Calculate user count at each step
step_1 = df['User_ID'].nunique()  # All users who signed up
step_2 = df[df['Trial_Usage_Level'].isin(['Medium', 'High'])]['User_ID'].nunique()
step_3 = df[df['Support_Used'] == 'Yes']['User_ID'].nunique()
step_4 = df[df['Bank_Connected'] == 'Yes']['User_ID'].nunique()
step_5 = df[df['Converted_to_Paid'] == 'Yes']['User_ID'].nunique()

# Build the funnel dataframe
funnel = pd.DataFrame({
    'Step': ['Signed Up', 'Used Features', 'Used Support', 'Connected Bank', 'Converted'],
    'Users': [step_1, step_2, step_3, step_4, step_5]
})

# Calculate % of users retained from previous step
funnel['Retained_from_previous (%)'] = funnel['Users'].div(funnel['Users'].shift(1)).round(2) * 100

# Calculate actual drop-off (%)
funnel['Dropoff_from_previous (%)'] = 100 - funnel["Retained_from_previous (%)"]

funnel
```

```python
## BEHAVIOUR  BASED CONVERSION FUNNEL

df['signed_up'] = True  # everyone signed up
df['used_features'] = df['Trial_Usage_Level'].notna()
df['used_support'] = df['Support_Used'] == 'Yes'
df['connected_bank'] = df['Bank_Connected'] == 'Yes'
df['converted'] = df['Converted_to_Paid'] == "Yes"
```

```python
## DEFINING USER BEHAVIOUR

### CREATING THE POWER USERS
(df['Converted_to_Paid'] == 'Yes') &(df["Trial_Usage_Level"]== "High")

## engaged but didnt convert(engaged - non converter
(df['Converted_to_Paid'] == 'No') & (df["Trial_Usage_Level"].isin(["Medium", "High"]))

## light users
(df['Converted_to_Paid'] == 'No') &(df["Trial_Usage_Level"]== "Low")  

## instant drop offs
(df['Converted_to_Paid'] == 'No') &(df["Trial_Usage_Level"]== "Low")  

## silent converters
(df['Converted_to_Paid'] == 'Yes') & (df['Support_Used'] =="No")

## support seekers
(df['Support_Used'] == 'Yes')
```

```python
## CREATING A NEW COLUMN USER_BEHAVIOR_BUCKET


def assign_user_bucket(row):
    if row['Converted_to_Paid'] == 'Yes' and row['Trial_Usage_Level'] == 'High':
        return 'Power User'
    elif row['Converted_to_Paid'] == 'No' and row['Trial_Usage_Level'] in ['Medium', 'High']:
        return 'Engaged Non-Converter'
    elif row['Converted_to_Paid'] == 'No' and row['Trial_Usage_Level'] == 'Low':
        return 'Light User'
    elif row['Converted_to_Paid'] == 'No' and row['Support_Used'] == 'No' and row['Trial_Usage_Level'] == 'Low':
        return 'Instant Drop-off'
    elif row['Converted_to_Paid'] == 'Yes' and row['Support_Used'] == 'No':
        return 'Silent Converter'
    elif row['Support_Used'] == 'Yes':
        return 'Support Seeker'
    else:
        return 'Other'

df['User_Behavior_Bucket'] = df.apply(assign_user_bucket, axis=1)
```

```python
df
```

```python
## creating a pivot table


bucket_summary = (
    df.groupby(['User_Behavior_Bucket', 'Converted_to_Paid'])
      .size()
      .reset_index(name='count')
)

# Pivot to make it easier to compare
pivot = bucket_summary.pivot(index='User_Behavior_Bucket',
                              columns='Converted_to_Paid',
                              values='count').fillna(0)

pivot['Total'] = pivot['Yes'] + pivot['No']
pivot['Conversion_Rate (%)'] = round((pivot['Yes'] / pivot['Total']) * 100, 1)

pivot = pivot.reset_index()
pivot
```
